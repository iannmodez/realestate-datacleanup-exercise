#### Exercise 09. Is the average of "Valdemorillo" and "Galapagar" prices the same? (★★☆)

Print the both average prices and then write a conclusion about them

```python
# TODO

import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

valdemorillo_ds = ds[ds['level5'] == 'Valdemorillo']['price']
galapagar_ds = ds[ds['level5'] == 'Galapagar']['price']

val_precio = valdemorillo_ds.mean()
gal_precio = galapagar_ds.mean()

print(f"Media de precios en Valdemorillo: {val_precio: .2f} euros")
print(f"Media de precios en Galapagar: {gal_precio: .2f} euros")

dif_max = 10000

if abs(val_precio - gal_precio) <= dif_max:
    print("Las medias de precios en Valdemorillo y Galapagar son esencialmente iguales.")
else:
    print("Las medias de precios en Valdemorillo y Galapagar no son iguales.")
```

#### Exercise 10. Is the average of "Valdemorillo" and "Galapagar" price per square meter (price/m2) the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
# TODO

import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds['price'] = ds['price'].replace((',','.'), '').astype(float)
ds['surface'] = ds['surface'].replace((',','.'), '').astype(float)

ds['pps'] = ds['price'] / ds['surface']

val_pps = ds[ds['level5'] == 'Valdemorillo']['pps']
gal_pps = ds[ds['level5'] == 'Galapagar']['pps']

val_pps_media = val_pps.mean()
gal_pps_media = gal_pps.mean()

print(f"Media de precio por metro cuadrado en Valdemorillo: {val_pps_media:.2f} euros/m2")
print(f"Media de precio por metro cuadrado en Galapagar: {gal_pps_media:.2f} euros/m2")

dif_max = 50

if abs(val_pps_media - gal_pps_media) <= dif_max:
    print("Las medias de precio por metro cuadrado en Valdemorillo y Galapagar son esencialmente iguales.")
else:
    if val_pps_media > gal_pps_media:
        print("El precio por metro cuadrado está mas caro en Valdemorillo")
    else:
        print("El precio por metro cuadrado esta mas caro en Galapar")
```

#### Exercise 11. Analyse the relation between the surface and the price of the houses (★★☆)

Hint: You can make a `scatter plot` and then write a conclusion about it

```python
# TODO: Code

import matplotlib.pyplot as plt
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

m_lim = 500  
ds = ds[ds['surface'] <= m_lim]

plt.figure(figsize=(12, 6))
plt.scatter(ds['surface'], ds['price'], alpha=0.6)

plt.title('Relación entre la Superficie y el Precio de las Viviendas (Superficie <= 500 m²)')
plt.xlabel('Superficie (m²)')
plt.ylabel('Precio (€)')
plt.grid(True)
plt.show()
```

Al analizar este gráfico se ha tenido que limitar la superficie a 500 metros cuadrados para poder visualizar correctamente la tendencia general, existe una gran concentración de viviendas en el rango de 20-30 metros a 200-230 metros y en el rango de precios de 0 a 1.000.000 de euros, la relación es a cuanta más superficie más caro es, sin embargo también existen viviendas mas caras a pesar de tener la misma superficie, esto indica que los metros cuadrados no son el único factor para que una vivienda sea más o menos cara.

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

for poblacion in poblaciones_ds['level5'].unique():
    subset = poblaciones_ds[poblaciones_ds['level5'] == poblacion]
    plt.hist(subset['price_normalized'], bins=30, alpha=0.3, histtype='stepfilled', 
             edgecolor='black', linewidth=1.2, label=poblacion)

plt.title('Histogramas Normalizados de Precios por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(title='Población')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO

import matplotlib.pyplot as plt

poblaciones_1 = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
p_fin = poblaciones_ds[poblaciones_ds['level5'].isin(poblaciones_1)]

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

for i, poblacion in enumerate(poblaciones_1):
    ax = axs[i//2, i%2]
    p_fin_fin = p_fin[p_fin['level5'] == poblacion]
    ax.scatter(p_fin_fin['surface'], p_fin_fin['price'], alpha=0.5, label=poblacion)
    ax.set_title(poblacion)
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio (€)')
    ax.legend()

plt.tight_layout()
plt.show()
```
